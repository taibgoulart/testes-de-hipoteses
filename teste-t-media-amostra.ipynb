Objetivo
Testar se a média do GPA é significativamente diferente de um valor específico (2.5).

Hipóteses
Hipótese nula (H₀): A média do GPA é 2.5.
Hipótese alternativa (H₁): A média do GPA é diferente de 2.5.

Resultados
t-statistic: -31.735
p-value: 9.46e-185
Interpretação
Com um p-valor extremamente pequeno (muito menor que 0.05), rejeitamos a hipótese nula (H₀). Isso indica que a média do GPA na amostra é significativamente diferente de 2.5.

Gráfico
O histograma mostra a distribuição do GPA com uma linha vertical vermelha representando a média esperada (2.5).

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = 'teste-sala.csv'
arquivo = pd.read_csv(dados)
```

Teste T para uma média da amostra

```python
media_gpa = 2.5
t_stat, p_value = stats.ttest_1samp(arquivo['GPA'], media_gpa)
```

```python
print(f'T-Satitistic: {t_stat}')
print(f'P-value: {p_value}')
```

Gráfico da Distribuição do GPA

```python
plt.figure(figsize=(10, 6))
sns.histplot(arquivo['GPA'], kde=True, bins=30)
plt.axvline(dados, color='r', linestyle='dashed', linewidth=2)
plt.title('Distribuição do GPA com a média de 2.5 esperada')
plt.xlabel('GPA')
plt.ylabel('Frequência')
plt.show()
```
