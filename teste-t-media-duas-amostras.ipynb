Objetivo
Comparar o GPA entre alunos que fazem tutorias e alunos que não fazem.

Hipóteses
Hipótese nula (H₀): As médias do GPA dos dois grupos são iguais.
Hipótese alternativa (H₁): As médias do GPA dos dois grupos são diferentes.

Resultados
t-statistic: -3.078
p-value: 0.0021
Interpretação
Com um p-valor menor que 0.05, rejeitamos a hipótese nula (H₀). Isso indica que há uma diferença significativa nas médias do GPA entre alunos que fazem tutorias e alunos que não fazem.

Gráfico
O boxplot compara a distribuição do GPA entre alunos que fazem tutoria e alunos que não fazem.

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = 'teste-sala.csv'
arquivo = pd.read_csv(dados)
```

```python
# Dividindo os dados em dois grupos
com_tutoria = arquivo[arquivo['Tutoring'] == 1]['GPA']
sem_tutoria = arquivo[arquivo['Tutoring'] == 0]['GPA']
```

```python
# Realizando teste para duas amostras independentes 
t_stat_ind, p_value_ind = stats.ttest_ind(com_tutoria, sem_tutoria)
```

```python
# mostrando os resultados 
print(f'T-Statistic: {t_stat_ind}')
print(f'P-Value: {p_value_ind}')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Tutoring', y='GPA', data=arquivo)
plt.title('Comparação do GPA entre alunos com e sem tutoria')
plt.xlabel('Tutoring (0 = Não, 1 = Sim)')
plt.ylabel('GPA')
plt.show()
```
