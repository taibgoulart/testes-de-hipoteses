Objetivo
Testar a independência entre duas variáveis categóricas: gênero e participação em atividades extracurriculares.

Hipóteses
Hipótese nula (H₀): As variáveis são independentes.
Hipótese alternativa (H₁): As variáveis não são independentes.

Resultados
Chi2 Statistic: 4.328
p-value: 0.115
Interpretação
Com um p-valor maior que 0.05, não rejeitamos a hipótese nula (H₀). Isso indica que não há evidência suficiente para afirmar que as variáveis gênero e participação em atividades extracurriculares são dependentes.

Gráfico
O gráfico de barras mostra a participação em atividades extracurriculares por gênero.

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
dados = 'teste-sala.csv'
arquivo = pd.read_csv(dados)
```

Gerando Tabela de Contigência 

```python
tabela_contigencia = pd.crosstab(arquivo['Gender'], arquivo['Extracurricular'])
```

```python
tabela_contigencia
```

Realizando Teste Qui-Quadrado


```python
chi2_stat, p_value, dof, expected = stats.chi2_contingency(tabela_contigencia)
```

```python
print(f'Chi2 Statistic: {chi2_stat}')
print(f'P-Value: {p_value}')
print(f'Degress of Freedom: {dof}')
print(f'Expected Frequencies: {expected}')
```

```python
tabela_contigencia.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Participação em Atividades Extracurriculares por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Frequência')
plt.show()
```
